```python
#Importing Useful Libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Creating a class which contains all Functions
class LogisticRegression(object):
    def __init__(self,alpha=0.3, n_iter=500):
        self.alpha = alpha
        self.n_iter = n_iter
        self.noi = []
        self.J_hist = []


    def _sigmoid(self, x):                                                      ##Creating a Sigmoid Function
     return 1 / (1 + np.exp(-x))-0.0000001

    def score(self, X, y):
     return sum(self.predict(X) == y) / len(y)

    def _predict_one(self, x):                                                  ##This function computes X.theta on each set of weights,takes the maximum value 
     return max((x.dot(theta),bias) for theta,bias in self.theta)[1]

    def predict(self, X):                                                       ##The values of above function are called here
     return [self._predict_one(i) for i in np.insert(X, 0, 1, axis=1)]          ##The insert function takes the arguments array name,index before whic value is inserted,value
    

    def fit(self, X, y):
        X = np.insert(X, 0, 1, axis=1)
        self.theta = []
        m = X.shape[0]
        for i in np.unique(y):
            y_copy = np.where(y == i, 1, 0)
            theta = np.ones(X.shape[1])

            for _ in range(self.n_iter):
                output = X.dot(theta)
                errors = y_copy - self._sigmoid(output)
                theta += self.alpha / m * errors.dot(X)
            self.theta.append((theta, i))
            self.errors = errors
            z = X@theta.T
          
            
            
        return self

        for i in range(self.n_iter):
         self.noi.append(i)
        
    def plotting(self):
      plt.plot(self.noi,self.J_hist)
      plt.xlabel("Number Of Iterations")
      plt.ylabel("Cost Function")
      plt.title("Cost Function vs Iteration")
      plt.show()
    
       
```
